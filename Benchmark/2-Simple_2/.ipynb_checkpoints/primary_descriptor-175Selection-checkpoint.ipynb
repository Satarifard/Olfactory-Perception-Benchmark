```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Builds PUBCHEM_CANONICAL_SMILES and OPTIONS for primary_descriptor-100.csv

- Input columns (required):
  question_ID, CAS number, compound.name, question_category, prompt.1, prompt.2,
  answer, Descriptor 2, Descriptor 3

- Output:
  primary_descriptor-100_with_smiles_options.csv
    â€˘ PUBCHEM_CANONICAL_SMILES  (PubChem CanonicalSMILES when possible;
      else RDKit-canonical fallback; else raw; else empty)
    â€˘ OPTIONS  (pipe-separated 4 unique choices: answer, Descriptor 2, Descriptor 3,
      + one random different answer from other rows; shuffled)

Progress is printed per row, including source notes.
"""

import csv
import random
import time
import urllib.parse
from typing import Optional, List, Set, Tuple

import pandas as pd
import requests

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ Config â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
IN_CSV  = "primary_descriptor.csv"
OUT_CSV = "primary_descriptor_with_smiles_options.csv"
SLEEP_BETWEEN_CALLS = 0.12  # be polite to public APIs

# HTTP session
SESSION = requests.Session()
SESSION.headers.update({
    "User-Agent": "SMILESFetcher/1.1 (local script for dataset prep)"
})
TIMEOUT = 15

PUBCHEM_BASE = "https://pubchem.ncbi.nlm.nih.gov/rest/pug"
OPSIN_BASE   = "https://opsin.ch.cam.ac.uk/opsin"

# Optional RDKit fallback
try:
    from rdkit import Chem
    from rdkit.Chem import MolToSmiles
    HAVE_RDKIT = True
except Exception:
    HAVE_RDKIT = False

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ HTTP helpers â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def _get_json(url: str, retries: int = 2, backoff: float = 0.6) -> Tuple[Optional[dict], int]:
    last_code = -1
    for i in range(retries + 1):
        try:
            r = SESSION.get(url, timeout=TIMEOUT)
            last_code = r.status_code
            if r.status_code == 200:
                return r.json(), r.status_code
            if r.status_code in (400, 404):
                return None, r.status_code
        except requests.RequestException:
            last_code = -1
        time.sleep(backoff * (2 ** i))
    return None, last_code

def _get_text(url: str, retries: int = 1, backoff: float = 0.6) -> Tuple[Optional[str], int]:
    last_code = -1
    for i in range(retries + 1):
        try:
            r = SESSION.get(url, timeout=TIMEOUT)
            last_code = r.status_code
            if r.status_code == 200:
                return r.text.strip(), r.status_code
            if r.status_code in (400, 404):
                return None, r.status_code
        except requests.RequestException:
            last_code = -1
        time.sleep(backoff * (2 ** i))
    return None, last_code

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ PubChem/OPSIN lookups â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def pc_smiles_from_RN(cas: str) -> Tuple[Optional[str], str]:
    if not cas or pd.isna(cas): return (None, "RN:empty")
    cas = str(cas).strip()
    if not cas: return (None, "RN:empty")
    url = f"{PUBCHEM_BASE}/compound/xref/RN/{urllib.parse.quote(cas)}/property/CanonicalSMILES/JSON"
    j, code = _get_json(url)
    if j:
        try:
            return j["PropertyTable"]["Properties"][0]["CanonicalSMILES"], "RNâ†’property 200"
        except Exception:
            return None, f"RNâ†’property parse-fail {code}"
    return None, f"RNâ†’property {code}"

def pc_smiles_from_name_property(name: str) -> Tuple[Optional[str], str]:
    if not name or pd.isna(name): return (None, "NAME:empty")
    name = str(name).strip()
    if not name: return (None, "NAME:empty")
    url = f"{PUBCHEM_BASE}/compound/name/{urllib.parse.quote(name)}/property/CanonicalSMILES/JSON"
    j, code = _get_json(url)
    if j:
        try:
            return j["PropertyTable"]["Properties"][0]["CanonicalSMILES"], "nameâ†’property 200"
        except Exception:
            return None, f"nameâ†’property parse-fail {code}"
    return None, f"nameâ†’property {code}"

def pc_smiles_via_cid(identifier: str) -> Tuple[Optional[str], str]:
    if not identifier or pd.isna(identifier): return (None, "CID:empty")
    ident = str(identifier).strip()
    if not ident: return (None, "CID:empty")
    url_cids = f"{PUBCHEM_BASE}/compound/name/{urllib.parse.quote(ident)}/cids/JSON"
    j, code = _get_json(url_cids)
    if not j:
        return None, f"nameâ†’cids {code}"
    try:
        cids = j["IdentifierList"]["CID"]
    except Exception:
        return None, f"nameâ†’cids parse-fail {code}"
    if not cids:
        return None, "nameâ†’cids empty"
    cid = cids[0]
    url_prop = f"{PUBCHEM_BASE}/compound/cid/{cid}/property/CanonicalSMILES/JSON"
    jp, code2 = _get_json(url_prop)
    if jp:
        try:
            return jp["PropertyTable"]["Properties"][0]["CanonicalSMILES"], f"cid({cid})â†’property 200"
        except Exception:
            return None, f"cid({cid})â†’property parse-fail {code2}"
    return None, f"cid({cid})â†’property {code2}"

def opsin_smiles(name: str) -> Tuple[Optional[str], str]:
    if not name or pd.isna(name): return (None, "OPSIN:empty")
    nm = str(name).strip()
    if not nm: return (None, "OPSIN:empty")
    url = f"{OPSIN_BASE}/{urllib.parse.quote(nm)}.smi"
    t, code = _get_text(url)
    if t and t.lower() != "null":
        # Return first token (SMILES) in case OPSIN adds name after a space
        return t.split()[0], "OPSIN 200"
    return None, f"OPSIN {code}"

def pubchem_canonicalize_from_smiles(smi: str) -> Tuple[Optional[str], str]:
    """Send any SMILES to PubChem and ask for CanonicalSMILES."""
    if not smi:
        return None, "canon:empty"
    url = f"{PUBCHEM_BASE}/compound/smiles/{urllib.parse.quote(smi)}/property/CanonicalSMILES/JSON"
    j, code = _get_json(url)
    if j:
        try:
            can = j["PropertyTable"]["Properties"][0]["CanonicalSMILES"]
            return can, "canon-from-smiles 200"
        except Exception:
            return None, f"canon-from-smiles parse-fail {code}"
    return None, f"canon-from-smiles {code}"

def rdkit_canonicalize(smi: str) -> Tuple[Optional[str], str]:
    if not HAVE_RDKIT or not smi:
        return None, "rdkit:unavailable-or-empty"
    try:
        mol = Chem.MolFromSmiles(smi)
        if mol is None:
            return None, "rdkit:parse-fail"
        return MolToSmiles(mol, canonical=True), "rdkit:canonical"
    except Exception:
        return None, "rdkit:error"

def lookup_pubchem_canonical_smiles(cas_number: str, compound_name: str) -> Tuple[Optional[str], str]:
    """Returns a PubChem-canonicalized SMILES if possible; includes provenance note."""
    notes = []

    # 1) Direct PubChem by RN
    smi, note = pc_smiles_from_RN(cas_number)
    notes.append(note)
    if smi:
        return smi, " | ".join(notes)

    # 2) PubChem by nameâ†’property
    smi, note = pc_smiles_from_name_property(compound_name)
    notes.append(note)
    if smi:
        return smi, " | ".join(notes)

    # 3) PubChem via CID (try name then CAS)
    for ident in (compound_name, cas_number):
        smi, note = pc_smiles_via_cid(ident)
        notes.append(note)
        if smi:
            return smi, " | ".join(notes)

    # 4) OPSIN (raw) â†’ attempt to canonicalize via PubChem
    smi, note = opsin_smiles(compound_name)
    notes.append(note)
    if smi:
        canon, cnote = pubchem_canonicalize_from_smiles(smi)
        notes.append(cnote)
        if canon:
            return canon, " | ".join(notes)
        # RDKit fallback if PubChem canon fails
        canon2, rnote = rdkit_canonicalize(smi)
        notes.append(rnote)
        if canon2:
            return canon2, " | ".join(notes)
        return smi, " | ".join(notes)  # raw OPSIN as last resort

    return None, " | ".join(notes)

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ OPTIONS builder â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def choose_distractors(
    base_items: List[str],
    answers_pool: List[str],
    extra_pool: List[str],
    target_k: int = 4,
) -> List[str]:
    clean = [str(x).strip() for x in base_items if isinstance(x, str) and str(x).strip()]
    unique: List[str] = []
    seen: Set[str] = set()
    for x in clean:
        if x not in seen:
            unique.append(x); seen.add(x)

    def add_from_pool(pool: List[str]):
        rnd = list({s.strip() for s in pool if isinstance(s, str) and s.strip() and s.strip() not in seen})
        random.shuffle(rnd)
        for s in rnd:
            if len(unique) >= target_k: break
            unique.append(s); seen.add(s)

    if len(unique) < target_k:
        add_from_pool(answers_pool)
    if len(unique) < target_k:
        add_from_pool(extra_pool)
    while len(unique) < target_k:
        filler = f"Option-{len(unique)+1}"
        if filler not in seen:
            unique.append(filler); seen.add(filler)
        else:
            unique.append(f"{filler}-{random.randint(1,999)}")

    random.shuffle(unique)
    return unique[:target_k]

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ Main â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def main():
    df = pd.read_csv(IN_CSV)

    required_cols = [
        "question_ID", "CAS number", "compound.name",
        "question_category", "prompt.1", "prompt.2",
        "answer", "Descriptor 2", "Descriptor 3"
    ]
    missing = [c for c in required_cols if c not in df.columns]
    if missing:
        raise SystemExit(f"âťŚ Missing required columns: {missing}")

    answers_all = df["answer"].dropna().astype(str).str.strip().tolist()
    d2_all = df["Descriptor 2"].dropna().astype(str).str.strip().tolist()
    d3_all = df["Descriptor 3"].dropna().astype(str).str.strip().tolist()

    smiles_cache = {}
    smiles_result: List[str] = []
    options_result: List[str] = []

    n = len(df)
    for i, row in df.iterrows():
        cas  = row.get("CAS number", "")
        name = row.get("compound.name", "")

        key = (str(cas).strip(), str(name).strip())
        if key in smiles_cache:
            smi, note = smiles_cache[key]
        else:
            smi, note = lookup_pubchem_canonical_smiles(cas, name)
            smiles_cache[key] = (smi, note)
            time.sleep(SLEEP_BETWEEN_CALLS)

        prefix = f"[{i+1}/{n}]"
        cas_s  = ("" if pd.isna(cas) else str(cas).strip())
        name_s = ("" if pd.isna(name) else str(name).strip())
        if smi:
            print(f"{prefix} âś… PUBCHEM_CANONICAL_SMILES: {smi}  || {note}  || CAS='{cas_s}'  NAME='{name_s}'", flush=True)
        else:
            print(f"{prefix} âťŚ SMILES NOT FOUND  || {note}  || CAS='{cas_s}'  NAME='{name_s}'", flush=True)

        smiles_result.append(smi if smi else "")

        # OPTIONS
        ans = row.get("answer", "")
        d2  = row.get("Descriptor 2", "")
        d3  = row.get("Descriptor 3", "")
        base_set = {str(x).strip() for x in [ans, d2, d3] if isinstance(x, str)}
        answers_pool = [a for a in answers_all if a and a not in base_set]
        extra_pool   = [x for x in (d2_all + d3_all) if x and x not in base_set]
        opts = choose_distractors([ans, d2, d3], answers_pool, extra_pool, target_k=4)
        options_result.append("|".join(opts))

    out_df = df.copy()
    # Insert directly after compound.name
    insert_at = out_df.columns.get_loc("compound.name") + 1
    out_df.insert(insert_at, "PUBCHEM_CANONICAL_SMILES", smiles_result)
    out_df["OPTIONS"] = options_result

    out_df.to_csv(OUT_CSV, index=False, quoting=csv.QUOTE_MINIMAL)
    print(f"\nâś… Done. Wrote: {OUT_CSV}")

if __name__ == "__main__":
    random.seed(42)
    main()
```
